**Desafio N° 1: Investigar por que o numero de experimentos com droga é muito maior do que os experimentos com controle.**





```python
dados_com_droga = dados[dados['tratamento'] == 'com_droga']
## exibe uma seleção dentro de uma seleção. Queremos criar um subconjunto de dados onde temos uma condição satisfeita.. nesse caso a condição é "com droga" ou "com controle"
## na coluna tratamento do nosso conj. de dados original, busque e salve em um outro conjunto de dados chamado "dados_com_droga", todos as linhas onte o valor da coluna
## tratamento é = a "com_droga"
```

```python
dados_com_droga
## exibe o subconjunto de dados que acabamos de criar, note que nele, a coluna tratamento apresenta apenas valores " com_droga" 
```

```python
dados_com_droga['composto'].value_counts()
## agrupando os tipos de compostos e somando sua incidência ( 718 experimentos receberam a droga do tipo 87d714366, 246 experimentos receberam a droga do tipo 9f80f3f77 ..... )
```

```python
  dados_com_controle = dados[dados['tratamento'] == 'com_controle']
  ## exibe uma seleção dentro de uma seleção. Queremos criar um subconjunto de dados onde temos uma condição satisfeita.. nesse caso a condição é "com droga" ou "com controle".
  ## na coluna tratamento do nosso conj. de dados original, busque e salve em um outro conjunto de dados chamado "dados_com_controle", todos as linhas onte o valor da coluna
  ## tratamento é = a "com_controle"
```

```python
dados_com_controle['composto'].value_counts()
## agrupando os tipos de compostos e somando sua incidência ( 1866 experimentos receberam o controle do tipo cacb2b860 ) ou seja, só existe um tipo de "controle" e este mesmo
## foi administrado em toda a população do experimento destinada a medicação de controle.
```

**Desafio 6: Melhorar a aparência dos gráficos e inserir labels**

```python
dados['tratamento'].value_counts().plot.bar()
## imprimindo um grafico de barras da maneira mais simples
```

```python
dados['dose'].value_counts().plot(kind='bar')
## uma outra forma de imprimir um grafico de barras ( um pouco mais complicada mas serve quando queremos colocar mais parâmetros na função plot, tipo tamanho do grafico por exemplo)
```

```python
dados['dose'].value_counts().plot(kind = 'bar', figsize = (10,5), title = 'DOSES ADMINISTRADAS', xlabel = 'tipo de dose', ylabel = 'amostras')
## Para aumentar o tamanho do grafico, inserir título e labels.  
```
